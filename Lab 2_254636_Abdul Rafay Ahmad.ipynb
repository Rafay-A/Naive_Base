## Task 2 ##

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'C:\Users\LENOVO\Desktop\15_monks-problems-2.csv', delimiter=";")
df.head(15)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

y = df['label']
X = df.filter(like='att')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder

o_enc = OrdinalEncoder()
o_enc.fit(df.drop(columns='label'))
X_train_e = o_enc.transform(X_train)
X_test_e = o_enc.transform(X_test)

l_enc = LabelEncoder()
l_enc.fit(df['label'])
y_train_e = l_enc.transform(y_train)
y_test_e = l_enc.transform(y_test)
```

```python
from sklearn.naive_bayes import CategoricalNB

clf = CategoricalNB()
clf.fit(X_train_e, y_train_e)

test_pred = clf.predict(X_test_e)
train_pred = clf.predict(X_train_e)
print("Test Score: ", accuracy_score(y_test_e, test_pred))
print("Train Score: ", accuracy_score(y_train_e, train_pred))
```

```python
import matplotlib.pyplot as plt

alpha_r = range(0,51)
test_er = []
train_er = []
for i in alpha_r:
    clf1 = CategoricalNB(alpha=i)
    clf1.fit(X_train_e, y_train_e)
    test_er.append(1- accuracy_score(y_test_e, clf1.predict(X_test_e)))
    train_er.append(1- accuracy_score(y_train_e, clf1.predict(X_train_e)))
plt.plot(alpha_r, test_er, label="test error", color="r")
plt.plot(alpha_r, train_er, label="train error", color="b")
plt.legend()
```

Best Value of Alpha: 15
